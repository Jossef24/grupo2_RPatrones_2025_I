```python
import pandas as pd
import sklearn
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression

from sklearn.metrics import r2_score, mean_squared_error, root_mean_squared_error, mean_absolute_error
from sklearn.pipeline import Pipeline

from sklearn.ensemble import RandomForestRegressor
```

```python
Y = pd.read_csv('FEB_catechol83.csv')
X = pd.read_csv('padel_83_1490.csv')
```

```python
X =  X.drop(['molecula'], axis=1)
X
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
pca = PCA(n_components = 0.9)
X_pca = pca.fit_transform(X_scaled)
X_processed = pd.DataFrame(X_pca)
X_processed
```

```python
Y =  Y.drop(['index_mol'], axis=1)
Y
```

```python
X_train,X_test,y_train,y_test=sklearn.model_selection.train_test_split(X_processed,Y,test_size=0.2,random_state=42)
```

```python
models = {
    'knn': {
        'model': Pipeline([('knn', KNeighborsRegressor())]),
        'params': {
            'knn__n_neighbors': [3, 5, 7, 9]
        }
    },
    'linear': {
        'model': Pipeline([('lr', LinearRegression())]),
        'params': {
            # Sin hiperparámetros que ajustar
        }
    },
    'random_forest': {
        'model': Pipeline([('rf', RandomForestRegressor(random_state=42))]),
        'params': {
            'rf__n_estimators': [50, 100],
            'rf__max_depth': [None, 5, 10]
        }
    }
}

results = {}
for name, mp in models.items():
    print(f"\n🏁 Entrenando modelo: {name}")
    gs = GridSearchCV(mp['model'], mp['params'], cv=5, scoring='r2', return_train_score=True, n_jobs=-1)
    gs.fit(X_train, y_train)
    results[name] = gs

    print("✅ Mejor puntuación (R² CV):", gs.best_score_)
    print("🔧 Mejores parámetros:", gs.best_params_)

    # Evaluación en conjunto de prueba
    y_pred = gs.predict(X_test)
    print("📊 R² en test:", r2_score(y_test, y_pred))
    print("📉 MSE en test:", mean_squared_error(y_test, y_pred))
    print("📉 RMSE en test:", root_mean_squared_error(y_test, y_pred))
    print("📉 MAE en test:", mean_absolute_error(y_test, y_pred))

```
